```
# funci칩n no recursiva. Se calcula con un bucle for..
def fib(n):
    "Compute the nth Fibonacci number"
    # Starting values for f0 and f1
    f0, f1 = 0, 1

    # Handle cases n==0 and n==1
    if n == 0:
        return 0
    elif n == 1:
        return 1
    
    # Start loop (from n = 2)    
    for i in range(2, n + 1):
        # Compute next term in sequence
        f = f1 + f0

        # Update f0 and f1    
        f0 = f1
        f1 = f

    # Return Fibonacci number
    return f

print(fib(10))
```

Dado que la secuencia de Fibonacci tiene una estructura recursiva, con el 洧녵췈 t칠rmino calculado a partir de los t칠rminos 洧녵-1 y 洧녵-2, podr칤amos escribir una funci칩n que utilice esta estructura recursiva:

```
# funci칩n recursiva.
def f(n): 
    "Compute the nth Fibonacci number using recursion"
    if n == 0:
        return 0  # This doesn't call f, so it breaks out of the recursion loop
    elif n == 1:
        return 1  # This doesn't call f, so it breaks out of the recursion loop
    else:
        return f(n - 1) + f(n - 2)  # This calls f for n-1 and n-2 (recursion), and returns the sum 

print(f(10))
```

```
from math import sqrt as square_root

def compute_roots(a, b, c):
    "Compute roots of the polynomial f(x) = ax^2 + bx + c"
    
    root0 = (-b + square_root(b*b - 4*a*c))/(2*a)
    root1 = (-b - square_root(b*b - 4*a*c))/(2*a)
    
    return root0, root1

# Compute roots of f = 4x^2 + 10x + 1
root0, root1 = compute_roots(4, 10, 1)
print(root0, root1)
```

### Ejemplo de rendimiento: c치lculo de la norma de un vector largo

La norma de un vector $x$ viene dada por: 

$$
\| x \| = \sqrt{\sum_{i=0}^{n-1} x_{i} x_{i}}
$$

donde $x_{i}$ es la entrada $i$ de $x$. Es el producto punto de un vector consigo mismo, 
seguido de la ra칤z cuadrada.
Para calcular la norma, podr칤amos hacer un bucle/iteraci칩n sobre las entradas del vector y sumar el cuadrado de cada entrada, y luego tomar la ra칤z cuadrada del resultado.

Vamos a evaluar la norma utilizando dos m칠todos para calcular la norma de una matriz de 10 millones de longitud para comparar su rendimiento. Primero creamos un vector con 10 millones de entradas aleatorias, utilizando NumPy:



```
# Create a NumPy array with 10 million random values
x = np.random.rand(10000000)
print(type(x))
```

Ahora cronometramos el tiempo que se tarda en calcular la norma del vector utilizando la funci칩n NumPy '`numpy.dot`'. Usamos el "comando m치gico" de Jupyter [`%time`](Notebook%20tips.ipynb#Simple-timing) para cronometrar la operaci칩n: 

```
%time norm = np.sqrt(np.dot(x, x))
print(norm)
```

La salida de tiempo de inter칠s es `Wall time`.

>  Usamos `%time` como una herramienta compacta y coveniente para 
> medir el tiempo que tarda en ejecutarse un comando.

Ahora realizamos la misma operaci칩n con nuestra propia funci칩n para calcular la norma:

```
def compute_norm(x):
    norm = 0.0
    for xi in x:
        norm += xi*xi
    return np.sqrt(norm)

%time norm =compute_norm(x)
print(norm)
```

Deber칤as ver que los dos enfoques dan el mismo resultado, pero la funci칩n 
NumPy es m치s de 100 veces m치s r치pida, 춰y posiblemente m치s de 100.000 veces m치s r치pida!

El mensaje es que las funciones y estructuras de datos especializadas para c치lculos num칠ricos pueden ser muchos 칩rdenes de magnitud m치s r치pidas que sus propias implementaciones generales. Adem치s, las funciones especializadas son mucho menos

### Operaciones con matrices en 2D

Para aquellos que hayan visto matrices anteriormente, las operaciones de esta secci칩n les resultar치n familiares. Para aquellos que no han visto las matrices, es posible que desee volver a esta secci칩n una vez que las matrices se han cubierto en las clases de matem치ticas.

#### Multiplicaci칩n matriz-vector y matriz-matriz

Consideraremos la matriz $A$:
$$
A  = 
\begin{bmatrix}
3.4 & 2.6 \\
2.1 & 4.5
\end{bmatrix}
$$

and the vector $x$:

$$
x  = 
\begin{bmatrix}
0.2 \\ -1.1
\end{bmatrix}
$$

```
A = np.array([[3.4, 2.6], [2.1, 4.5]])
print("Matrix A:\n {}".format(A))

x = np.array([0.2, -1.1])
print("Vector x:\n {}".format(x))
```

Podemos calcular el producto matriz-vector $y = Ax$ mediante:

```
y = A.dot(x)
print(y)
```

La multiplicaci칩n matriz-matriz se realiza de forma similar. Computar $C = AB$, donde $A$, $B$ y $C$ son matrices:

```
B = np.array([[1.3, 0], [0, 2.0]])

C = A.dot(B)
print(C)
```

The inverse of a matrix ($A^{-1}$) and the determinant ($\det(A)$) can be computed using functions in the NumPy submodule `linalg`:

# Extracci칩n de matrices (Array Slicing)

Cuando se trabaja con matrices, a menudo es 칰til extraer subconjuntos de una matriz. Podemos querer s칩lo las 3 primeras entradas de un array largo, o podemos querer la segunda columna de un array 2D (matriz). Estas operaciones se conocen como *array slicing* (https://en.wikipedia.org/wiki/Array_slicing).

Exploraremos el slicing a trav칠s de ejemplos. Comenzamos creando un array lleno de valores aleatorios:

```
x = np.random.rand(5)
print(x)
```

A continuaci칩n se muestran algunas operaciones de corte:

```
# Using ':' implies the whole range of indices, i.e. from 0 -> (length-1)
y = x[:]
print("Slice using '[:]' {}".format(y))

# Using '1:3' implies indices 1 -> 3 (not including 3)
y = x[1:3]
print("Slice using '[1:3]': {}".format(y))

# Using '2:-1' implies indices 2 -> second-from-last
y = x[2:-1]
print("Slice using '[2:-1]': {}".format(y))

# Using '2:-2' implies indices 2 -> third-from-last
y = x[2:-2]
print("Slice using '[2:-2]': {}".format(y))
```

> N칩tese el uso del 칤ndice `-1`. El 칤ndice `-1` corresponde a la 칰ltima entrada de la matriz, y `-2` a la 
> la pen칰ltima entrada, etc. Esto es conveniente si sabemos a qu칠 distancia del final de un array se encuentra una entrada deseada.
> Usando 칤ndices negativos podemos expresar esto sin referencia a la longitud del array.

Si queremos que un slice vaya desde el principio de un array, o hasta el final de un array, lo hacemos: 

```
# Using ':3' implies start -> 3 (not including 3)
y = x[:3]
print("Slice using '[:3]': {}".format(y))

# Using '4:' implies 4 -> end
y = x[4:]
print("Slice using '[4:]': {}".format(y))

# Using ':' implies start -> end
y = x[:]
print("Slice using '[:]': {}".format(y))
```

El corte puede aplicarse a matrices 2D:

```
B = np.array([[1.3, 0], [0, 2.0]])
print(B)

# Extract second row
r = B[1, :]
print(r)
```

# Ejemplo de m칩dulo: procesamiento en paralelo

Los m칩dulos est치ndar pueden simplificar problemas muy t칠cnicos. Un ejemplo es el procesamiento paralelo.

La mayor칤a de las CPUs -desde los tel칠fonos hasta los superordenadores- tienen ahora CPUs con m칰ltiples n칰cleos, y cada n칰cleo realiza c치lculos. Para beneficiarse de los m칰ltiples n칰cleos, tenemos que calcular en *paralelo*.
Un programa "est치ndar" realiza las tareas en orden, y en este caso s칩lo se utilizar치 un n칰cleo y el resto permanecer치
permanecer치 inactivo. Para obtener el mejor rendimiento del hardware, tenemos que calcular en paralelo. Es decir, realizar varias tareas al mismo tiempo. 

El procesamiento paralelo es un tema enorme en s칤 mismo, pero podemos tocarlo aqu칤 porque tenemos bibliotecas est치ndar
que facilitan su uso. La gesti칩n de tareas paralelas a bajo nivel es extremadamente t칠cnica, pero las bibliotecas est치ndar pueden facilitarla. Utilizaremos el m칩dulo `multiprocessing`, y lo usaremos para ordenar listas de n칰meros de forma concurrente.

Empezaremos por ver c칩mo generar una lista de enteros aleatorios utilizando el m칩dulo `random`. El siguiente c칩digo 
crea una lista (m치s sobre listas en el siguiente cuaderno) de 10 enteros aleatorios en el rango de 0 a 100 (sin incluir el 100):

```
import random
x = random.sample(range(0, 100), 10)
print(x)
```

Para crear una lista ordenada, utilizamos la funci칩n incorporada `sorted`:

```
y = sorted(x)
print(y)
```

Ahora, si necesitamos ordenar varias listas diferentes, podr칤amos ordenar las listas una tras otra, o podr칤amos ordenar varias listas al mismo tiempo (en paralelo). Nuestro sistema operativo gestionar치 entonces el env칤o de la tarea de ordenaci칩n a los diferentes n칰cleos del procesador. Antes de ver c칩mo hacer esto, implementamos una funci칩n para realizar la ordenaci칩n:

```
import multiprocessing
import random

def mysort(N):
    "Create a list of random numbers of length N, and return a sorted list"

    # Create random list 
    x = random.sample(range(0, N), N) 

    # Print process identifier (just out of interest)
    print("Process id: {}".format(multiprocessing.current_process()))
    
    # Return sorted list of numbers
    return sorted(x)
```

Para crear las listas ordenadas, poniendo a disposici칩n tres procesos (hilos), utilizamos:

```
N = 20000
with multiprocessing.Pool(processes=3) as p:
    p.map(mysort, [N, N, N])  # Call function mysort three times
```

Vemos en la salida que tres procesos diferentes han trabajado en nuestro problema - uno para cada tarea de clasificaci칩n.

Utilizamos el procesamiento en paralelo para hacer los c치lculos m치s r치pidos. Vamos a cronometrar nuestro c치lculo utilizando diferentes n칰meros de procesos para ver c칩mo afecta al rendimiento. Para realizar el cronometraje, primero encapsulamos nuestro problema en una funci칩n:

```
def parallel_sort(N, num_proc):
    "Create three lists of random numbers (each of length N) using num_proc processes"
    with multiprocessing.Pool(processes=num_proc) as p:
        p.map(mysort, [N, N, N])
```

Usando el comando '%time', cronometramos la ordenaci칩n usando s칩lo un proceso (el 칰nico proceso ordena las listas una tras otra):

```
N = 500000
%time parallel_sort(N, 1)    
```

Vemos por el `Process id` que tres procesos diferentes han trabajado en las listas. La ejecuci칩n en paralelo deber칤a ser m치s r치pida (esto depender치 del hardware).
