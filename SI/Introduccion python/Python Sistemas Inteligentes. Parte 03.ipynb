```
# función no recursiva. Se calcula con un bucle for..
def fib(n):
    "Compute the nth Fibonacci number"
    # Starting values for f0 and f1
    f0, f1 = 0, 1

    # Handle cases n==0 and n==1
    if n == 0:
        return 0
    elif n == 1:
        return 1
    
    # Start loop (from n = 2)    
    for i in range(2, n + 1):
        # Compute next term in sequence
        f = f1 + f0

        # Update f0 and f1    
        f0 = f1
        f1 = f

    # Return Fibonacci number
    return f

print(fib(10))
```

Dado que la secuencia de Fibonacci tiene una estructura recursiva, con el 𝑛º término calculado a partir de los términos 𝑛-1 y 𝑛-2, podríamos escribir una función que utilice esta estructura recursiva:

```
# función recursiva.
def f(n): 
    "Compute the nth Fibonacci number using recursion"
    if n == 0:
        return 0  # This doesn't call f, so it breaks out of the recursion loop
    elif n == 1:
        return 1  # This doesn't call f, so it breaks out of the recursion loop
    else:
        return f(n - 1) + f(n - 2)  # This calls f for n-1 and n-2 (recursion), and returns the sum 

print(f(10))
```

# Usando funciones de la biblioteca: Ejemplo de math

Para utilizar una función de un módulo tenemos que hacerla disponible en nuestro programa. 
Esto se llama "importar". Lo hemos hecho en cuadernos anteriores con el módulo `math`, pero sin explicación. El proceso se explica a continuación.

El módulo `math` (https://docs.python.org/3/library/math.html) proporciona una amplia gama de funciones matemáticas. Por ejemplo, para calcular la raíz cuadrada de un número, hacemos:

```
import math

x = 2.0
x = math.sqrt(x)
print(x)
```

Al prefijar `sqrt` con `math`, estamos utilizando un *espacio de nombres (namespace)* (que en este caso es `math`).
Esto deja claro qué función `sqrt` queremos usar - podría haber más de una función `sqrt` disponible.

>*Namespace:* El prefijo `math` indica qué función `sqrt` queremos utilizar. Este
puede parecer pedante, pero en la práctica a menudo hay diferentes algoritmos para realizar la misma o similar operación. Pueden variar en velocidad y precisión. En algunas aplicaciones podemos necesitar un método preciso (pero lento) para calcular la raíz cuadrada, mientras que para otras aplicaciones podemos necesitar velocidad con un compromiso en la precisión. Pero, si dos funciones tienen el mismo nombre y no se distinguen por un espacio de nombres, tenemos un *choque de nombres*.

> En un programa grande, dos desarrolladores pueden elegir el mismo nombre para dos funciones que realizan tareas similares pero ligeramente diferentes. Si estas funciones están en diferentes módulos, no habrá un choque de nombres ya que el nombre del módulo proporciona un "espacio de nombres" - un prefijo que proporciona una distinción entre las dos funciones. Los espacios de nombres son muy útiles para los programas de varios autores. Uno de los puntos débiles de los lenguajes más antiguos, como C y Fortran, es que no admiten espacios de nombres. La mayoría de los lenguajes modernos sí soportan los espacios de nombres.  

Podemos importar funciones específicas de un módulo, por ejemplo, importar sólo la función `sqrt`:

```
from math import sqrt as some_math_function 

x = 2.0
x = some_math_function(x)
print(x)
```

Renombrar las funciones en la importación puede ser útil para mantener el código corto, y veremos más adelante que puede ser útil para cambiar entre diferentes funciones.
Sin embargo, la elección del nombre de arriba es una práctica muy pobre - el nombre 'función_matemática` no es descriptivo.
A continuación se muestra un ejemplo más sensato.

```
from math import sqrt as square_root

def compute_roots(a, b, c):
    "Compute roots of the polynomial f(x) = ax^2 + bx + c"
    
    root0 = (-b + square_root(b*b - 4*a*c))/(2*a)
    root1 = (-b - square_root(b*b - 4*a*c))/(2*a)
    
    return root0, root1

# Compute roots of f = 4x^2 + 10x + 1
root0, root1 = compute_roots(4, 10, 1)
print(root0, root1)
```

### Ejemplo de rendimiento: cálculo de la norma de un vector largo

La norma de un vector $x$ viene dada por: 

$$
\| x \| = \sqrt{\sum_{i=0}^{n-1} x_{i} x_{i}}
$$

donde $x_{i}$ es la entrada $i$ de $x$. Es el producto punto de un vector consigo mismo, 
seguido de la raíz cuadrada.
Para calcular la norma, podríamos hacer un bucle/iteración sobre las entradas del vector y sumar el cuadrado de cada entrada, y luego tomar la raíz cuadrada del resultado.

Vamos a evaluar la norma utilizando dos métodos para calcular la norma de una matriz de 10 millones de longitud para comparar su rendimiento. Primero creamos un vector con 10 millones de entradas aleatorias, utilizando NumPy:



```
# Create a NumPy array with 10 million random values
x = np.random.rand(10000000)
print(type(x))
```

Ahora cronometramos el tiempo que se tarda en calcular la norma del vector utilizando la función NumPy '`numpy.dot`'. Usamos el "comando mágico" de Jupyter [`%time`](Notebook%20tips.ipynb#Simple-timing) para cronometrar la operación: 

```
%time norm = np.sqrt(np.dot(x, x))
print(norm)
```

La salida de tiempo de interés es `Wall time`.

>  Usamos `%time` como una herramienta compacta y coveniente para 
> medir el tiempo que tarda en ejecutarse un comando.

Ahora realizamos la misma operación con nuestra propia función para calcular la norma:

```
def compute_norm(x):
    norm = 0.0
    for xi in x:
        norm += xi*xi
    return np.sqrt(norm)

%time norm =compute_norm(x)
print(norm)
```

Deberías ver que los dos enfoques dan el mismo resultado, pero la función 
NumPy es más de 100 veces más rápida, ¡y posiblemente más de 100.000 veces más rápida!

El mensaje es que las funciones y estructuras de datos especializadas para cálculos numéricos pueden ser muchos órdenes de magnitud más rápidas que sus propias implementaciones generales. Además, las funciones especializadas son mucho menos 
de tener errores.

#### Multiplicación matriz-vector y matriz-matriz

Consideraremos la matriz $A$:
$$
A  = 
\begin{bmatrix}
3.4 & 2.6 \\
2.1 & 4.5
\end{bmatrix}
$$

and the vector $x$:

$$
x  = 
\begin{bmatrix}
0.2 \\ -1.1
\end{bmatrix}
$$

```
A = np.array([[3.4, 2.6], [2.1, 4.5]])
print("Matrix A:\n {}".format(A))

x = np.array([0.2, -1.1])
print("Vector x:\n {}".format(x))
```

Podemos calcular el producto matriz-vector $y = Ax$ mediante:

```
y = A.dot(x)
print(y)
```

La multiplicación matriz-matriz se realiza de forma similar. Computar $C = AB$, donde $A$, $B$ y $C$ son matrices:

```
B = np.array([[1.3, 0], [0, 2.0]])

C = A.dot(B)
print(C)
```

The inverse of a matrix ($A^{-1}$) and the determinant ($\det(A)$) can be computed using functions in the NumPy submodule `linalg`:

```
Ainv = np.linalg.inv(A)
print("Inverse of A:\n {}".format(Ainv))

Adet = np.linalg.det(A)
print("Determinant of A: {}".format(Adet))
```

# Extracción de matrices (Array Slicing)

Cuando se trabaja con matrices, a menudo es útil extraer subconjuntos de una matriz. Podemos querer sólo las 3 primeras entradas de un array largo, o podemos querer la segunda columna de un array 2D (matriz). Estas operaciones se conocen como *array slicing* (https://en.wikipedia.org/wiki/Array_slicing).

Exploraremos el slicing a través de ejemplos. Comenzamos creando un array lleno de valores aleatorios:

```
x = np.random.rand(5)
print(x)
```

A continuación se muestran algunas operaciones de corte:

```
# Using ':' implies the whole range of indices, i.e. from 0 -> (length-1)
y = x[:]
print("Slice using '[:]' {}".format(y))

# Using '1:3' implies indices 1 -> 3 (not including 3)
y = x[1:3]
print("Slice using '[1:3]': {}".format(y))

# Using '2:-1' implies indices 2 -> second-from-last
y = x[2:-1]
print("Slice using '[2:-1]': {}".format(y))

# Using '2:-2' implies indices 2 -> third-from-last
y = x[2:-2]
print("Slice using '[2:-2]': {}".format(y))
```

> Nótese el uso del índice `-1`. El índice `-1` corresponde a la última entrada de la matriz, y `-2` a la 
> la penúltima entrada, etc. Esto es conveniente si sabemos a qué distancia del final de un array se encuentra una entrada deseada.
> Usando índices negativos podemos expresar esto sin referencia a la longitud del array.

Si queremos que un slice vaya desde el principio de un array, o hasta el final de un array, lo hacemos: 

```
# Using ':3' implies start -> 3 (not including 3)
y = x[:3]
print("Slice using '[:3]': {}".format(y))

# Using '4:' implies 4 -> end
y = x[4:]
print("Slice using '[4:]': {}".format(y))

# Using ':' implies start -> end
y = x[:]
print("Slice using '[:]': {}".format(y))
```

El corte puede aplicarse a matrices 2D:

```
B = np.array([[1.3, 0], [0, 2.0]])
print(B)

# Extract second row
r = B[1, :]
print(r)
```

El método __init__ es un método especial de una clase en Python. El objetivo fundamental del método __init__ es inicializar los atributos del objeto que creamos.

El método __init__ es el primer método que se ejecuta cuando se crea un objeto.
El método __init__ se llama automáticamente.

```
class Empleado:

    def __init__(self):
        self.nombre=input("Ingrese el nombre del empleado:")
        self.sueldo=float(input("Ingrese el sueldo:"))

    def imprimir(self):
        print("Nombre:",self.nombre)
        print("Sueldo:",self.sueldo)

    def paga_impuestos(self):
        if self.sueldo>3000:
            print("Debe pagar impuestos")
        else:
            print("No paga impuestos")


# bloque principal

empleado1=Empleado()
empleado1.imprimir()


```

```
empleado1=Empleado()
```

```
empleado1.imprimir()
```

```
empleado1.paga_impuestos()
```

Desarrollar una clase que represente un punto en el plano y tenga los siguientes métodos: inicializar los valores de x e y que llegan como parámetros, imprimir en que cuadrante se encuentra dicho punto (concepto matemático, primer cuadrante si x e y son positivas, si x<0 e y>0 segundo cuadrante, etc.)

```
class Punto:

    def __init__(self,x,y):
        self.x=x
        self.y=y

    def imprimir(self):
        print("Coordenada del punto")
        print("(",self.x,",",self.y,")")

    def imprimir_cuadrante(self):
        if self.x>0 and self.y>0:
            print("Primer cuadrange")
        else:
            if self.x<0 and self.y>0:
                print("Segundo cuadrante")
            else:
                if self.x<0 and self.y<0:
                    print("Tercer cuadrante")
                else:
                    if self.x>0 and self.y<0:
                        print("Cuarto cuadrante")

```

```
punto1=Punto(10,-2)
punto1.imprimir()
punto1.imprimir_cuadrante()
```

## Método especial __str__

Podemos hacer que se ejecute un método definido por nosotros cuando pasamos un objeto a la función print o cuando llamamos a la función str (convertir a string)

¿Qué sucede cuando llamamos a la función print y le pasamos como parámetro un objeto?



```
class Persona:
    def __init__(self,nom,ape):
        self.nombre=nom
        self.apellido=ape

    def __str__(self):
        cadena=self.nombre+","+self.apellido
        return cadena

```

```
persona1=Persona("Jose","Rodriguez")
persona2=Persona("Ana","Martinez")
print(str(persona1)+"-"+str(persona2))  # Jose,Rodriguez-Ana,Martinez 
```

```
print(persona1)
```

Crearle el método especial __str__ para retornar un string con el formato (x,y).

```
class Punto:

    def __init__(self, x, y):
        self.x=x
        self.y=y

    def __str__(self):
        return "("+str(self.x)+","+str(self.y)+")"
```

```
punto1=Punto(10,3)
punto2=Punto(3,4)
print(punto1)
print(punto2)
```

```
class Familia:

    def __init__(self,padre,madre,hijos=[]):
        self.padre=padre
        self.madre=madre
        self.hijos=hijos

    def __str__(self):
        cadena=self.padre+","+self.madre
        for hi in self.hijos:
            cadena=cadena+","+hi
        return cadena


# bloque principal

familia1=Familia("Pablo","Ana",["Pepe","Julio"])
familia2=Familia("Jorge","Carla")
familia3=Familia("Luis","Maria",["marcos"])

print(familia1)
print(familia2)
print(familia3)
```

# Ejemplo de módulo: procesamiento en paralelo

Los módulos estándar pueden simplificar problemas muy técnicos. Un ejemplo es el procesamiento paralelo.

La mayoría de las CPUs -desde los teléfonos hasta los superordenadores- tienen ahora CPUs con múltiples núcleos, y cada núcleo realiza cálculos. Para beneficiarse de los múltiples núcleos, tenemos que calcular en *paralelo*.
Un programa "estándar" realiza las tareas en orden, y en este caso sólo se utilizará un núcleo y el resto permanecerá
permanecerá inactivo. Para obtener el mejor rendimiento del hardware, tenemos que calcular en paralelo. Es decir, realizar varias tareas al mismo tiempo. 

El procesamiento paralelo es un tema enorme en sí mismo, pero podemos tocarlo aquí porque tenemos bibliotecas estándar
que facilitan su uso. La gestión de tareas paralelas a bajo nivel es extremadamente técnica, pero las bibliotecas estándar pueden facilitarla. Utilizaremos el módulo `multiprocessing`, y lo usaremos para ordenar listas de números de forma concurrente.

Empezaremos por ver cómo generar una lista de enteros aleatorios utilizando el módulo `random`. El siguiente código 
crea una lista (más sobre listas en el siguiente cuaderno) de 10 enteros aleatorios en el rango de 0 a 100 (sin incluir el 100):

```
import random
x = random.sample(range(0, 100), 10)
print(x)
```

Para crear una lista ordenada, utilizamos la función incorporada `sorted`:

```
y = sorted(x)
print(y)
```

Ahora, si necesitamos ordenar varias listas diferentes, podríamos ordenar las listas una tras otra, o podríamos ordenar varias listas al mismo tiempo (en paralelo). Nuestro sistema operativo gestionará entonces el envío de la tarea de ordenación a los diferentes núcleos del procesador. Antes de ver cómo hacer esto, implementamos una función para realizar la ordenación:

```
import multiprocessing
import random

def mysort(N):
    "Create a list of random numbers of length N, and return a sorted list"

    # Create random list 
    x = random.sample(range(0, N), N) 

    # Print process identifier (just out of interest)
    print("Process id: {}".format(multiprocessing.current_process()))
    
    # Return sorted list of numbers
    return sorted(x)
```

Para crear las listas ordenadas, poniendo a disposición tres procesos (hilos), utilizamos:

```
N = 20000
with multiprocessing.Pool(processes=3) as p:
    p.map(mysort, [N, N, N])  # Call function mysort three times
```

Vemos en la salida que tres procesos diferentes han trabajado en nuestro problema - uno para cada tarea de clasificación.

Utilizamos el procesamiento en paralelo para hacer los cálculos más rápidos. Vamos a cronometrar nuestro cálculo utilizando diferentes números de procesos para ver cómo afecta al rendimiento. Para realizar el cronometraje, primero encapsulamos nuestro problema en una función:

```
def parallel_sort(N, num_proc):
    "Create three lists of random numbers (each of length N) using num_proc processes"
    with multiprocessing.Pool(processes=num_proc) as p:
        p.map(mysort, [N, N, N])
```

Usando el comando '%time', cronometramos la ordenación usando sólo un proceso (el único proceso ordena las listas una tras otra):

```
N = 500000
%time parallel_sort(N, 1)    
```

Vemos por el `Process id` que tres procesos diferentes han trabajado en las listas. La ejecución en paralelo debería ser más rápida (esto dependerá del hardware).
