###La sentencia If###

En el ejemplo, la sentencia de control comienza con un if, seguido de la expresi칩n a comprobar, seguida de ':'

Debajo hay un bloque de c칩digo, con cuatro espacios de sangr칤a, que se ejecuta si la comprobaci칩n (x > 0,0) es verdadera:

```
x = -10.0  # Initial x value

if x > 0.0:  
    print('Initial x is greater than zero')
    x -= 20.0
elif x < 0.0:  
    print('Initial x is less than zero')
    x += 21.0
else: 
    print('Initial x is not less than zero and not greater than zero, therefore it must be zero')
    x *= 2.5

# Print new x value
print("New x value:", x)
```

```
N = 50  # Check numbers up 50 for primes (excludes 50)

# Loop over all numbers from 2 to 50 (excluding 50)
for n in range(2, N):

    # Assume that n is prime
    n_is_prime = True
    # Check if n can be divided by m, where m ranges from 2 to n (excluding n)
    for m in range(2, n):
         if n % m == 0:   # Esto es cierto si el resto de n/m es igual a cero
                          # Hemos comprobado que n es divisible por m, por lo que no puede ser un n칰mero primo. 
                          # No es necesario comprobar m치s valores de m, as칤 que establece n_es_primo = Falso y
                          # salimos del bucle 'm'.
            n_is_prime = False
            break

    #  If n is prime, print to screen        
    if n_is_prime:
        print(n)
```

Bucles infinitos: causa y protecci칩n
Un error com칰n, especialmente cuando se utilizan sentencias while, es el bucle infinito. Esto ocurre cuando se introduce un bucle pero nunca termina (sale). Los bucles infinitos pueden hacer que un sistema no responda, a veces requiriendo un apagado para restaurar la funci칩n.

Es una buena pr치ctica, especialmente cuando se est치 aprendiendo, a침adir protecciones contra los bucles infinitos. Por ejemplo,

```
x = 0.0

counter = 0
while x < 0.05:

    # Guard against infinite loop
    counter += 1
    if counter > 2000:
        print("Loop count exceeded 2000. Exiting")
        break
```

Ejemplo: representaci칩n inexacta.
Es trivial que

Si 洧논=0,1 , podemos escribir

El error de redondeo para el caso de 0,1 es peque침o, y en muchos casos no supondr치 un problema. Sin embargo, a veces los errores de redondeo pueden acumularse y destruir la precisi칩n:
Ahora, partiendo de 洧논=0,1 evaluamos el lado derecho para obtener un 'nuevo' 洧논 , y utilizamos este nuevo 洧논 para evaluar de nuevo el lado derecho. La aritm칠tica es trivial: 洧논 debe seguir siendo igual a 0,1 . Probamos esto en un programa que repite este proceso 20 veces:

```
x = 0.1
for i in range(20):
    x = x*11 - 1
    print(x)
```

La soluci칩n estalla y se desv칤a ampliamente de 洧논=0,1 . Los errores de redondeo se amplifican en cada paso, lo que lleva a una respuesta completamente err칩nea. La representaci칩n inform치tica de 0,1 no es exacta, y cada vez que multiplicamos 0,1 por 11 , aumentamos el error en un factor de 10 aproximadamente (podemos ver arriba que perdemos un d칤gito de precisi칩n en cada paso). Se puede observar el mismo problema utilizando programas de hojas de c치lculo.

Si utilizamos 洧논=0,5 , que se puede representar exactamente en binario:

```
x = 0.5
for i in range(20):
    x = x*11 - 5
    print(x)
```

El resultado es exacto en este caso.

Por defecto, Python utiliza 64 bits para almacenar un float. Podemos utilizar el m칩dulo NumPy para crear un float que utilice s칩lo 32 bits. Probando esto para el caso 洧논=0.1:

```
import numpy as np
x = np.float32(0.1)
for i in range(20):
    x = x*11 - 1
    print(x)
```

El error se dispara m치s r치pido en este caso comparado con el de 64 bits - usar 32 bits lleva a una peor aproximaci칩n de 0,1 que cuando se usan 64 bits.
